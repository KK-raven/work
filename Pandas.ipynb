```python
df = pd.read_csv('tmdb_5000_movies.csv')

df['budget'] = df['budget'].apply(lambda x: np.nan if x == 0 else x)
df['revenue'] = df['revenue'].apply(lambda x: np.nan if x == 0 else x)
df = df.dropna(subset=['budget', 'revenue'])
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
df['profit'].sort_values(ascending=False)
```

```python
df = df.sort_values(by='profit', ascending=False)
# df = df.reset_index(drop=True)
df
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')

df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
df = df.sort_values('profit', ascending=False)
df.to_csv('movie_sorted_profit.csv', index=False)

pd.read_csv('movie_sorted_profit.csv')
```
